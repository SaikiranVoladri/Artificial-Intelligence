# Simple Linear Regression 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Libraries

## Importing Data Set

```python
data_set=pd.read_csv('Salary.csv')
x=data_set.iloc[:,: -1].values#Slicing all rows ,All columns Excet last column
y=data_set.iloc[:,-1].values# All rows and only lastcolumn
# print(x,y)
 
# X=data_set['years_experience'].values
# print(data_set)
```

## Splitting into test and training set

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test=train_test_split(x, y ,test_size=0.4,random_state=42)# testsize=0.2 ==(20 % for testing)and 80% for training , randomstate is to ensure that everytime every time same amount of testdata is splitted
# print(y_test)
```

## Training on Training Set using Simple Linear Regression


```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train , y_train)
```

## Prediction on Test Set

```python
y_pred = regressor.predict(x_test)
print(y_pred)
```

## Visualizing the Training Set

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color='green')
plt.plot(x_train , regressor.predict(x_train))
plt.title("Linear regression")
plt.xlabel("EXPERIENCE")
plt.ylabel=("CTC")
plt.show()
```

## Visualizing the Test Set

```python
import matplotlib.pyplot as plt
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train , regressor.predict(x_train))
plt.title("Linear regression")
plt.xlabel("EXPERIENCE")
plt.ylabel=("CTC")
plt.show()
```
