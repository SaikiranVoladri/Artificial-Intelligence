```python
l5 =[[i+j for j in range(3)] for i in range(5)]
print(l5)
```

```python
l6=[[3*i+j for j in range(3)] for i in range(5)]
print(l6)
```

#Taking input from user

```python
r= int(input("enter rows"))
c=int(input('enter columns'))
l1=[ int(x) for x in input().split()] # l1=[ int(x) for x in input()]
print("list is ",l1)

a=[]
for i in range(r):
  a.append([])

  for  j in range(c):
    a[i].append(l1[c*i+j])
print("updated list",a)
```

#Matplotlib -> for graphical representation

```python
x1=[1,2,3,4]
x2=[8,9,7,6]
y1=[4,3,2,1]
y2=[4,3,2,1]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```
